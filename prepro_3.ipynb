### We're missing movies from 2018 - 2021. Thus, we'll scrape wikipedia to supplement our dataset

```python
link_2018 = "https://en.wikipedia.org/wiki/List_of_American_films_of_2018"
link_2020 = "https://en.wikipedia.org/wiki/List_of_American_films_of_2020"
```

```python
source = urllib.request.urlopen(link_2018).read()
soup = bs.BeautifulSoup(source, "lxml")
```

```python
tables = soup.find_all("table", class_="wikitable sortable")
tables
```

```python
len(tables)
```

#### The movies have been devided into 4 tables (according to months):
- Januar - March
- April - June
- July - September
- October - December

#### Thus, we need to group them up into 1 dataset

```python
#In order to convert it into a Dataframe instead of a list, we add [0] at the end of pd.read_html
df1 = pd.read_html(str(tables[0]))[0]
df2 = pd.read_html(str(tables[1]))[0]
df3 = pd.read_html(str(tables[2]))[0]
df4 = pd.read_html(str(tables[3]))[0]
```

```python
df = df1.append(df2.append(df3.append(df4, ignore_index=True), ignore_index=True), ignore_index=True)
df
```

```python
feature_list = ["Title", "Cast and crew"]
```

```python
df_2018 = df[feature_list]
df_2018
```
